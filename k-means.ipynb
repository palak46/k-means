```python
import numpy as np 
import sklearn 
import matplotlib.pyplot as plt
```

```python
x=np.random.uniform(1,15,size=100)
y=np.random.uniform(1,15,size=100)
```

```python
x0=np.random.uniform(17,30,size=150)
y0=np.random.uniform(17,30,size=150)
```

```python
x1=np.random.uniform(20,30,size=100)
y1=np.random.uniform(1,15,size=100)
```

```python
r_n=np.random.uniform(0,30,size=15)
r_n1=np.random.uniform(0,30,size=15)
```

```python
x=np.concatenate((x,x0,x1,r_n), axis=None)
y=np.concatenate((y,y0,y1,r_n1), axis=None)
```

```python
plt.scatter(x,y)
```

```python
t1=np.random.uniform(0,30,size=1)
t2=np.random.uniform(0,30,size=1)
t3=np.random.uniform(0,30,size=1)

p1=np.random.uniform(0,30,size=1)
p2=np.random.uniform(0,30,size=1)
p3=np.random.uniform(0,30,size=1)

t=np.concatenate((t1,t2,t3), axis=None)
p=np.concatenate((p1,p2,p3), axis=None)

plt.scatter(x,y)
plt.scatter(t,p)

```

```python
def distance(t,x,p,y):
    return np.sqrt((t-x)**2+(p-y)**2)
```

```python
for i in range(1):
    
    
    d0=distance(t[0],x,p[0],y)
    d1=distance(t[1],x,p[1],y)
    d2=distance(t[2],x,p[2],y)
    d0=d0.reshape(d0.shape[0],1)
    d1=d1.reshape(d1.shape[0],1)
    d2=d2.reshape(d2.shape[0],1)
    distance1=np.concatenate((d0,d1,d2),axis=1)
    mylist=[]
    for i in range(len(distance1)):
        min_value = np.min(distance1[i])
        for g,h in enumerate(distance1[i]):
            if(min_value==h):
                mylist.append(g)
    l0=[]
    l1=[]
    l2=[]

    o0=[]
    o1=[]
    o2=[]

    for i in range(len(mylist)):
        if(mylist[i]==0):
            l0.append(x[i])
            o0.append(y[i])
        if(mylist[i]==1):
            l1.append(x[i])
            o1.append(y[i])
        if(mylist[i]==2):
            l2.append(x[i])
            o2.append(y[i])
    t[0]=np.mean(l0)
    t[1]=np.mean(l1)
    t[2]=np.mean(l2)
    p[0]=np.mean(o0)
    p[1]=np.mean(o1)
    p[2]=np.mean(o2)
```

```python
plt.scatter(l0,o0)
plt.scatter(l1,o1)
plt.scatter(l2,o2)
```
